```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
import spacy
from matplotlib.pyplot import imread
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline
```

```python
reviews = pd.read_csv('iphone_4s_review.csv',encoding='Latin-1')
reviews
```

```python
reviews['Reviews'] = reviews['Reviews'].astype(str)
```

```python
reviews = [word.strip() for word in reviews['Reviews']] 
reviews = [word for word in reviews if word]
reviews
```

```python
review_text = ' '.join(reviews)
review_text[0:1000]
```

```python
no_punc_text=review_text.translate(str.maketrans('','',string.punctuation))
no_punc_text[0:1000]
```

```python
from nltk.tokenize import word_tokenize
text_tokens=word_tokenize(no_punc_text)
print(text_tokens[0:1000])
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
my_stop_words=stopwords.words('english')

sw_list = ['rt','ye','yeah','haha','Yes','I']
my_stop_words.extend(sw_list)

no_stop_tokens=[word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens[0:1000])
```

```python
# Normalize the data
lower_words=[Text.lower() for Text in no_stop_tokens]
print(lower_words[100:200])
```

```python
# Stemming (Optional)
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmed_tokens=[ps.stem(word) for word in lower_words]
print(stemmed_tokens[100:200])
```

```python
clean_tweets=' '.join(stemmed_tokens)
clean_tweets[0:1000]
```
